```python
def categorize_age (Age):
    if Age < 18:
        return 'Alaikainen'
    elif Age < 60:
        return 'Aikuinen'
    else:
        return 'Seniori'

data['Ikatyypit'] = data['Age'].apply(categorize_age)
data.head()
```
